```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import warnings
warnings.filterwarnings("ignore")
```

```python
#Load and View Data
```

```python
data_1 = pd.read_csv('c1data.csv')
```

```python
data_1.info()
```

```python
data_1.Drug.hist(bins=30)
plt.xlabel('Drug')
plt.ylabel('count')
```

```python
#View relationships in catergorical data
```

```python
plt.pie(x=data_1['Sex'].value_counts().values, labels=data_1['Sex'].value_counts().index, autopct='%1.1f%%')
```

```python
plt.pie(x=data_1['BP'].value_counts().values, labels=data_1['BP'].value_counts().index, autopct='%1.1f%%')
```

```python
plt.pie(x=data_1['Cholesterol'].value_counts().values, labels=data_1['Cholesterol'].value_counts().index, autopct='%1.1f%%')
```

```python
#Analysis of relationships with Drug data
```

```python
sns.scatterplot(data=data_1, x="Drug", y="Na_to_K")
plt.show()
```

```python
sns.swarmplot(data=data_1,
            x='Cholesterol',
            y='Drug')
plt.xlabel('Cholesterol')
plt.ylabel('Drug')
sns.despine()
plt.show()
```

```python
#View Relationship of blood pressure and Na to K
```

```python
sns.swarmplot(data=data_1,
            x='BP',
            y='Na_to_K')
plt.xlabel('BP')
plt.ylabel('Na_to_K')
sns.despine()
plt.show()
```

```python
# Hot Encode Catergorical Features
```

```python
ohed = pd.get_dummies(data_1, columns = ['Sex', 'BP', 'Cholesterol', 'Drug']) 
print(ohed)
```

```python
sns.scatterplot(data=ohed, x="Drug_drugY", y="Na_to_K")
plt.show()
```

```python
sns.scatterplot(data=ohed, x="Drug_drugX", y="Na_to_K")
plt.show()
```
